To perform time series analysis on the data, we must change the index from a range index to a time series index. In the cell below, create a time series index using the `pd.date_range` function. Our next step is to decompose the time series and evaluate the patterns in the data. Load the `statsmodels.api` submodule and plot the decomposed plot of `userAcceleration.x`. Set `freq=60` in the `seasonal_decompose` function. Your graph should look like the one below.

![time series decomposition](../images/tsa_decompose.png)

```python
# Your code here:

import statsmodels.api as sm
import matplotlib.pyplot as plt
acceleration = sm.tsa.seasonal_decompose(sensor['userAcceleration.x'], freq=60, model='additive')
rotation = sm.tsa.seasonal_decompose(sensor['rotationRate.x'], freq=60, model='additive')
acceleration.plot()
plt.title('User Acceleration')
rotation.plot()
plt.title('Rotation Rate')
plt.show()
```

Plot the decomposed time series of `rotationRate.x` also with a frequency of 60.

The next step will be to test both variables for stationarity. Perform the Augmented Dickey Fuller test on both variables below.

```python
# Your code here:

from statsmodels.tsa.stattools import adfuller

#Perform Dickey-Fuller test:
def fuller_test(timeseries):
    print('Results of Dickey-Fuller Test for timeseries')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used', 'Critical Value'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    return dfoutput

print(fuller_test(sensor['userAcceleration.x']))
print('\n')
print(fuller_test(sensor['rotationRate.x']))
```

What are your conclusions from this test?

```python
# Your conclusions here:

La hipótesis nula se rechaza en ambas series, lo cual indica que la série temporal es estacionaria, 
es decir la media y la varianza son constantes en el tiempo y la autocovarianza no depende del tiempo.
```
